```python
%reset -f

import numpy as np
pi = np.pi
```

```python
# attached plating thickness
t = 0.016
# plate width
b = 0.62372
# plate length
L = 2.94

# 模量和泊松比
E = 207e9
v = 0.3

Iyy = 0.015 * 0.2**3 / 3
Izz = 0.2 * 0.015**3 / 12

Z_bar = 0.1

J = 0.2 * 0.015**3 / 3


k = E * t**3 / 3 / b / (1 - v**2)
Io = Iyy + Izz
G = E / 2 / (1 + v)
```

```python
""" Simply Supported Stiffener Ends """

# 最小临界应力在 m, m>1
m = L / pi * (k / E / Izz / Z_bar**2)**(1/4)

sigma_cr = (E * Izz * Z_bar**2 * (m * pi / L)**2 + G * J + k * (L/m/pi)**2) / Io
print(sigma_cr/10**6)
```

```python
""" Built-in stiffener ends without warping """

# 最小临界应力在 m, m>1
m = L / 2 / pi * (3 * k / E / Izz / Z_bar**2)**(1/4)

sigma_cr = (16 * E * Izz * Z_bar**2 * (m * pi / L)**2 + 4 * G * J + 3 * k * (L/m/pi)**2) / Io / 4
print(sigma_cr)
```
