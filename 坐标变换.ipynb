```python
import numpy as np

def transf(theta, n_nodes=2, degree=False):
    """
    生成二维杆/梁单元的坐标变换矩阵
    
    参数
    ----
    theta : float
        单元局部 x 轴相对于全局 x 轴的夹角
    n_nodes : int
        节点数
    degree : bool
        如果 theta 使用角度制，则设为 True；默认使用弧度制
    
    返回
    ----
    t : ndarray
        尺寸为 (2*n_nodes, 2*n_nodes) 的坐标变换矩阵
    """
    
    if degree:
        theta = np.deg2rad(theta)
    
    c = np.cos(theta)
    s = np.sin(theta)
    
    t_node = np.array([[ c, s],
                       [-s, c]])
    
    t = np.kron(np.eye(n_nodes), t_node)
    return t
```
