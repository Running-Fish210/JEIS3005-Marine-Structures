```python
%reset -f

import numpy as np
pi = np.pi
deg = np.degrees
sin = np.sin
cos = np.cos
tan = np.tan

arcsin = np.arcsin
arccos = np.arccos
arctan = np.arctan
```

```python

def section(dp, tp, tw, dw, df, tf):
    """
    This function used to calcult section parameter
    The bottom of plate is x-axis
    
    dp: 平板长
    tp: 平板厚
    tw: web 厚
    dw: web 长
    df: flange 宽
    tf flange 厚
    """
    ap = tp * dp
    aw = tw * dw
    af = tf * df

    a = ap + aw + af

    na = (ap * tp/2 + aw * (tp + dw/2) + af * (tp + dw + tf/2)) / (a)
    ip = dp * tp**3 / 12
    iw = tw * dw**3 / 12
    i_f = df * tf**3 / 12

    i = ip + ap * (tp/2 - na)**2 + iw + aw * (tp + dw/2 - na)**2 + i_f + af * (tp + dw + tf/2 - na)**2
    ym = tp + dw + tf - na

    z = i / ym

    eaa = a / 2 / dp
    zp = a / 2 * eaa / 2 + aw * (tp - eaa + dw/2) + af * (tp - eaa + dw + tf/2) + a / 2 * (tp - eaa)/2

    phi = zp / z

    # result = np.array([ap, aw, af, a, NA, ip, iw, if, I_NA, z, zp, phi])
    result = np.array([ap, aw, af, a, na, ip, iw, i_f, i, z, zp, phi])
    return result
```

```python
r = section(700e-3, 12e-3, 13e-3, 140e-3, 60e-3, 60e-3)
print(f'Ap = {r[0]}')
print(f'Aw = {r[1]}')
print(f'Af = {r[2]}')
print(f'A = {r[3]}')

print(f'NA = {r[4]}')

print(f'I_NA = {r[8]}')

print(f'Z = {r[9]}')
print(f'Zp = {r[10]}')
print(f'phi = {r[11]}')

```
